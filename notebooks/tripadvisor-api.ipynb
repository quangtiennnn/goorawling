```python
import os
import requests
import pandas as pd

def fetch_tripadvisor_details_and_save(df, n, api_key, province_name="Ho Chi Minh"):
    ids = df.id[:n]  # Get first n restaurant_id (assuming index is restaurant_id)
    records = []
    headers = {"accept": "application/json"}

    for location_id in ids:
        url = f"https://api.content.tripadvisor.com/api/v1/location/{location_id}/details?key={api_key}&language=en&currency=USD"
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            data = response.json()
            data['restaurant_id'] = location_id
            records.append(data)
        else:
            print(f"Failed for id {location_id}: {response.status_code}")

    # Save to CSV
    output_dir = os.path.join("data", "tripadvisor_restaurant")
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, f"{province_name}.csv")
    pd.DataFrame(records).set_index("restaurant_id").to_csv(output_path, encoding="utf-8-sig")
    print(f"Saved {len(records)} records to {output_path}")

# Example usage:
fetch_tripadvisor_details_and_save(df, 3000, tripadvisor_api_key, "Ha Noi")
```
