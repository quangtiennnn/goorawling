```python
import streamlit as st
import pandas as pd

# Load your DataFrame
df = pd.read_csv("data/tripadvisor_restaurant/restaurant_links.csv")
df = df[0:4] # Limit to first 4 rows for demonstration

st.title("Browse and Input DataFrame Rows")

# Show the DataFrame
st.dataframe(df)

st.write("## Input for Each Row")

for idx, row in df.iterrows():
    with st.form(key=f"form_{idx}"):
        st.write(f"**Row {idx}**")
        st.write(row)
        # Example input: you can add more fields as needed
        user_input = st.text_input(f"Input for restaurant_id {row['restaurant_id']}", key=f"input_{idx}")
        submit = st.form_submit_button("Submit")
        if submit:
            st.success(f"Received input for {row['restaurant_id']}: {user_input}")
```
